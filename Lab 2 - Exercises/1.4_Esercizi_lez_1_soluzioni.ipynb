### **1. Trasformare cicli in list comprehension**

Fai tre esempi di cicli producono una lista e poi riscrivi lo stesso codice usando la list comprehension.

```python
# Esempio 1: quadrati dei numeri da 0 a 5
quadrati_loop = []
for numero in range(6):
    quadrati_loop.append(numero ** 2)

quadrati_comprehension = [numero ** 2 for numero in range(6)]

# Esempio 2: numeri pari da 0 a 10
pari_loop = []
for numero in range(11):
    if numero % 2 == 0:
        pari_loop.append(numero)

pari_comprehension = [numero for numero in range(11) if numero % 2 == 0]

# Esempio 3: parole trasformate in maiuscolo
parole = ["python", "numpy", "array"]

maiuscole_loop = []
for parola in parole:
    maiuscole_loop.append(parola.upper())

maiuscole_comprehension = [parola.upper() for parola in parole]

print("Quadrati con ciclo:", quadrati_loop)
print("Quadrati con list comprehension:", quadrati_comprehension)
print("Pari con ciclo:", pari_loop)
print("Pari con list comprehension:", pari_comprehension)
print("Maiuscole con ciclo:", maiuscole_loop)
print("Maiuscole con list comprehension:", maiuscole_comprehension)
```

### **2. Vettore di Numeri Primi**

1. **Crea un vettore** contenente tutti i numeri primi compresi tra **0 e 10**  
   _(Puoi scriverli direttamente nell'array)._

2. **Conta quanti numeri ci sono nel vettore** utilizzando la funzione `len()`.  **Ottieni lo stesso numero** accedendo all'attributo `.size` del vettore.

3. Quale pensi sia il tipo di dato (`dtype`) del vettore?  Prova a rispondere senza eseguire il codice e Verifica la tua risposta accedendo all'attributo `.dtype` del vettore

4. Scrivi l'array usando una list comprehension che controlla che i numeri siano primi. 

```python
import numpy as np

numeri_primi = np.array([2, 3, 5, 7])

print("Vettore numeri primi:", numeri_primi)
print("Numero di elementi con len():", len(numeri_primi))
print("Numero di elementi con .size:", numeri_primi.size)
print("Tipo di dato del vettore:", numeri_primi.dtype)

def is_prime(number):
    if number < 2:
        return False

    for divisor in range(2, number):
        if number % divisor == 0:
            return False

    return True

numeri_primi_comprehension = np.array([number for number in range(11) if is_prime(number)])

print("Vettore creato con list comprehension:", numeri_primi_comprehension)
```

### **3.  Operazioni su array**
Crea un array **a**, 1D (senza digitarlo esplicitamente), e genera due nuovi array **b**, contenente una sottostringa a piacere e **c** che è il reverse dell'array **a** (esempio [1,2,3] diventa [3,2,1]) .  Dividi l'array **a** per l'array **c**. Fai la stessa cosa per una lista.

```python
a = np.arange(1, 11)
b = a[2:7]
c = a[::-1]

divisione_array = a / c

print("Array a:", a)
print("Array b:", b)
print("Array c:", c)
print("Divisione a / c:", divisione_array)

lista_a = list(range(1, 11))
lista_b = lista_a[2:7]
lista_c = lista_a[::-1]
divisione_liste = [value_a / value_c for value_a, value_c in zip(lista_a, lista_c)]

print("Lista a:", lista_a)
print("Lista b:", lista_b)
print("Lista c:", lista_c)
print("Divisione lista_a / lista_c:", divisione_liste)
```
